# Janet Benchmarks Game

Compare Janet to Python and Lua for some of the Computer Language Benchmarks Game
"artificial" benchmarks.  See all the caveats at: https://benchmarksgame-team.pages.debian.net/benchmarksgame/index.html

See the plot below.  Janet beats python typically, and loses to lua typically,
but there are exceptions.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import io
def getstats(fpath):
    tag = "BENCHXXX: "
    datalines = [line[len(tag):]
                 for line in open(fpath)
                 if line.startswith(tag)]
    inp = io.StringIO("".join(datalines))
    df = pd.read_csv(inp)
    return df

df = getstats("bench.log")
df           
```

```python
df.plot.bar(x='name', figsize=(20,10))
```
